```python
category_df = df_cleaned['Category'].value_counts()[:10]
category_df
```

```python
category_df = df_cleaned.groupby('Category').agg({'Installs': pd.Series.sum})
```

```python
category_df.sort_values('Installs', ascending=True, inplace=True)
```

```python
bar2 = px.bar(x=category_df['Installs'], y=category_df.index)
```

```python
bar2.show()
```
